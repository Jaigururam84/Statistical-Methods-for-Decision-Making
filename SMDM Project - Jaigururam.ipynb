```python
#Libaries needed
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#checking for missing values
missing_values = data.isnull().sum()
print(missing_values)
```

```python
#we found gender and partner salary have missing values and there data type are object and float

#For numerical columns (Partner salary) we are going to Impute missing values with the mean.
ncolumns = data.select_dtypes(include=['float64', 'int64']).columns
data[ncolumns] = data[ncolumns].fillna(data[ncolumns].mean())
#For categorical columns (gender) we are going to Impute missing values with the Mode.
ccolumns = data.select_dtypes(include=['object']).columns
data[ccolumns] = data[ccolumns].fillna(data[ccolumns].mode().iloc[0])
```

```python
Q1 = data['No_of_Dependents'].quantile(0.25)
Q3 = data['No_of_Dependents'].quantile(0.75)
IQR = Q3 - Q1

LB = Q1 - (1.5 * IQR)
UB = Q3 + (1.5 * IQR)

outliers = data[(data['No_of_Dependents'] < LB) | (data['No_of_Dependents'] > UB)]

print(outliers)
```

```python
#observation
#outlier are valid data no need to change/treat it.
```

```python
#1. Do men tend to prefer SUVs more compared to women?

#Answere : No Male do not prefer SUV compared to female.From below plot we can see that Female Owns SUV more compare to Male
```

```python
crosstab = pd.crosstab(data['Gender'], data['Make'] == 'SUV')
crosstab['SUV_Percentage'] = crosstab[True] / crosstab.sum(axis=1)
print(crosstab)

crosstab['SUV_Percentage'].plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Percentage of SUV Owners')
plt.title('SUV Ownership by Gender')
plt.show()
```

```python
#2. What is the likelihood of a salaried person buying a Sedan?

#Answere : The likelihood of a salaried person buying a Sedan is: 44.2%
```

```python
salaried_individuals = data[data['Profession'] == 'Salaried']
num_salaried_sedan_owners = salaried_individuals[salaried_individuals['Make'] == 'Sedan'].shape[0]
num_salaried_individuals = salaried_individuals.shape[0]
likelihood = num_salaried_sedan_owners / num_salaried_individuals
print("The likelihood of a salaried person buying a Sedan is:", likelihood*100,"Percentage")
```

```python
#3. What evidence or data supports Sheldon Cooper's claim that a salaried male is an easier target for a SUV sale over a Sedan sale?

# Answere : The data does not support Sheldon Cooper's claim. As Percentage of salaried female owning SUV was more compared to male
        #Percentage of salaried males who own an SUV: 13.39%
        #Percentage of salaried females who own an SUV: 52.67%
```

```python
salaried_males = data[(data['Profession'] == 'Salaried') & (data['Gender'] == 'Male')]
num_salaried_males_suv_owners = salaried_males[salaried_males['Make'] == 'SUV'].shape[0]
num_salaried_males = salaried_males.shape[0]
percentage_salaried_males_suv_owners = (num_salaried_males_suv_owners / num_salaried_males) * 100

salaried_females = data[(data['Profession'] == 'Salaried') & (data['Gender'] == 'Female')]
num_salaried_females_suv_owners = salaried_females[salaried_females['Make'] == 'SUV'].shape[0]
num_salaried_females = salaried_females.shape[0]
percentage_salaried_females_suv_owners = (num_salaried_females_suv_owners / num_salaried_females) * 100

print("Percentage of salaried males who own an SUV:", percentage_salaried_males_suv_owners)
print("Percentage of salaried females who own an SUV:", percentage_salaried_females_suv_owners)

if percentage_salaried_males_suv_owners > percentage_salaried_females_suv_owners:
  print("The data supports Sheldon Cooper's claim that a salaried male is an easier target for a SUV sale over a Sedan sale.")
else:
  print("The data does not support Sheldon Cooper's claim.")
```
